## Preprocessing

```python
import os
from os.path import dirname

path_labels = './labels'

labelfiles = [(dirname(os.path.join(root, name)).split("\\")[-1], name)
             for root, dirs, files in os.walk(path_labels)
             for name in files
             if name.endswith(".lab")]
```

```python
print len(labelfiles)
```

```python
# load lab files as csv (with tab separator)
import pandas as pd

df_labels = None

for lab in labelfiles:
    df = pd.read_csv(os.path.join(path_labels, *lab), sep='\t', header=None, names=['start', 'end', 'chord'])

    # add column with song index
    df['song_ix'] = lab[0]
    
    if df_labels is None:
        df_labels = df
    else:
        df_labels = df_labels.append(df)
```
